```python
# Collect the monthly G
import networkx as nx
def G_period(period):
    d = Data.loc[Data['period']==period]
    d = d.reset_index(drop=True)
    part_1 = Data.loc[Data['Year'] < d['Year'][0]]
    part_2 = Data.loc[(Data['Year']==d['Year'][0]) & (Data['month']<=d['month'][0])]
    Exsit_author =pd.concat([part_1,part_2])
    Exsit_author = Exsit_author.reset_index(drop=True)
    
    a = []
    for i in range(len(Exsit_author['Authors'])):
        a.append(Exsit_author['Authors'][i].split('-')[0].lower().replace(',',''))
    
    df = [[]] * len(d['Authors'])
    for i in range(len(d['Authors'])):
        df[i] = [d['Authors'][i].split('-')[0].lower().replace(',','')] + d['References'][i][1:][:-1].split(', ')
    
    GG = nx.DiGraph()
    
    for i in range(len(d['Authors'])):
        rough_author = df[i]
        for k in range(1,len(rough_author)):
            # edge = rough_author[0],rough_author[x]
            if rough_author[k] in a:
                attr = GG.get_edge_data(rough_author[0],rough_author[k])
                if attr != None:
                    attr = GG.get_edge_data(rough_author[0],rough_author[k])['weight'] + 1
                    GG.add_weighted_edges_from([(rough_author[0],rough_author[k],attr)])
                else:
                    GG.add_weighted_edges_from([(rough_author[0],rough_author[k],1)])
    if '[anonymous]' in GG.nodes():
        GG.remove_node('[anonymous]')
    
    # add node attribute of interval
    dict_node = defaultdict(float)
    dict_edge = defaultdict(float)
    dict_w = defaultdict(float)
    
    for i in range(len(GG)):
        dict_node[GG.nodes()[i]] = '<[%s,%s]>' %(period,period)
    nx.set_node_attributes(GG, 'Timestamp', dict_node)
    
    for i in range(len(GG.edges())):
        dict_edge[GG.edges()[i]] = '<[%s,%s]>' %(period,period)
        w = GG.get_edge_data(GG.edges()[i][0],GG.edges()[i][1])['weight']
        dict_w[GG.edges()[i]] = '<[%s,%s,%s]>' %(period,period,w)
    nx.set_edge_attributes(GG, 'Timestamp', dict_edge)
    nx.set_edge_attributes(GG, 'tw', dict_w)
    
    
    return GG
```

```python
# merge the total 18 graph
def merge_dynamic(Dynamic_G,G):
    # add nodes
    exsit_n = Dynamic_G.nodes()
    extend_n = G.nodes()
    for i in range(len(extend_n)):
        node = extend_n[i]
        if node in exsit_n:
            Dynamic_G.node[node]['Timestamp'] = Dynamic_G.node[node]['Timestamp'].replace('>',';')\
                                                + G.node[node]['Timestamp'].replace('<','')
        else:
            Dynamic_G.add_node(node)
            Dynamic_G.node[node] = G.node[node]
    # add edges
    exsit_ed = Dynamic_G.edges()
    extend_ed = G.edges()
    for i in range(len(extend_ed)):
        edge = extend_ed[i]
        if edge in exsit_ed:
            Dynamic_G.edge[edge[0]][edge[1]]['Timestamp'] = Dynamic_G.edge[edge[0]][edge[1]]['Timestamp'].replace('>',';')\
                                                                + G.edge[edge[0]][edge[1]]['Timestamp'].replace('<','')
            Dynamic_G.edge[edge[0]][edge[1]]['tw'] = Dynamic_G.edge[edge[0]][edge[1]]['tw'].replace('>',';')\
                                                                + G.edge[edge[0]][edge[1]]['tw'].replace('<','')
            Dynamic_G.edge[edge[0]][edge[1]]['weight'] = G.edge[edge[0]][edge[1]]['weight']
        else:
            Dynamic_G.add_edge(edge[0],edge[1])
            Dynamic_G.edge[edge[0]][edge[1]] = G.edge[edge[0]][edge[1]]
            
    return Dynamic_G
               
```

```python
tf.reset_default_graph()
S0 = Input(shape=[3,1])  
S1 = Input(shape=[3,1]) 
S2 = Input(shape=[3,1]) 
h0 = LSTM(32, activation='relu',dropout=0.2, recurrent_dropout=0.2, input_shape=(3,1))(S0)
h1 = LSTM(32, activation='relu',dropout=0.2, recurrent_dropout=0.2, input_shape=(3,1))(S1)
h2 = LSTM(32, activation='relu',dropout=0.2, recurrent_dropout=0.2, input_shape=(3,1))(S2)
#h1 = Dense(HIDDEN2_UNITS, activation='relu')(h0)
Centrality = Dense(1,activation='sigmoid',input_shape=(32,1))(h0)  
Pagerank = Dense(1,activation='sigmoid',input_shape=(32,1))(h1)   
Number = Dense(1,activation='sigmoid',input_shape=(32,1))(h2) 
V = Concatenate()([Centrality,Pagerank,Number])
output = Dense(1,activation='sigmoid')(V)
model = Model(inputs=[S0,S1,S2],outputs=output)

graph = tf.get_default_graph()
```
